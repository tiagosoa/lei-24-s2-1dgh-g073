The US10 main objective is to analyze the usage of the many park equipments and to help the Green Space Manager know which equipments are the favorites of the users. 

## Code and results
---

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

EQUIPMENT_USED_PATH = "EquipmentUsed.csv"

# Load the data
data = pd.read_csv(EQUIPMENT_USED_PATH)

# Calculate the counts and percentages of equipment usage
equipment_counts = data['Equipment'].value_counts()
total_users = len(data)
equipment_percentages = (equipment_counts / total_users) * 100

# Create a pie chart to visualize the data
plt.pie(equipment_percentages, labels=equipment_percentages.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('muted'))
plt.title('Percentage Use of Park Equipment')
plt.axis('equal')
plt.show()
```

## INTRODUCTION 
User Story 9 aims to analyze water consumption costs in parks through statistical methods, including monthly consumption barplots, average costs per park, and detailed comparisons of statistical indicators between parks with highest and lowest water usage.

## Code and results

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("water_consumption.csv", sep=";", decimal=",")

def filter_data(data, year, start_month, end_month, park):
    filtered_data = data[(data["Year"] == year) &
                         (data["Month"].between(start_month, end_month)) &
                         (data["Park"] == park)]
    return filtered_data

# Sample user specifications
year = 2023
start_month = 1  # January
end_month = 12    # March
park_id = 'Oriental'

filtered_data = filter_data(data, year, start_month, end_month, park_id)

# Step 3: Group the filtered data by month and aggregate the water consumption
monthly_consumption = filtered_data.groupby('Month')['Consumption'].sum().reset_index()

# Step 4: Plot the aggregated water consumption for each month using Seaborn
sns.barplot(x='Month', y='Consumption', data=monthly_consumption)
plt.title('Monthly Water Consumption')
plt.xlabel('Month')
plt.ylabel('Water Consumption')
plt.grid(True)
plt.show()

def calculate_monthly_average(data, parks_to_analyze):
    # Filtra os dados para os parques especificados pelo usuário
    filtered_data2 = data[data['Park'].isin(parks_to_analyze)]

    # Agrupa os dados por parque e mês
    grouped_data = filtered_data2.groupby(['Park', 'Month'])['Consumption'].mean().reset_index()

    # Renomeia a coluna para 'Average Consumption'
    grouped_data.rename(columns={'Consumption': 'Average Consumption'}, inplace=True)

    return grouped_data

# Solicita ao usuário para inserir os parques que deseja analisar
parks_to_analyze = input("Por favor, insira os parques que deseja calcular a média, separados por vírgula: ").split(',')

# Utiliza a função para calcular a média mensal
average_consumption = calculate_monthly_average(data, parks_to_analyze)

# Exibe os resultados
print(average_consumption)





```

## Analysis and Interpretation of the Results 
The goal of this analysis is to calculate and visualize the monthly water consumption for the Oriental park in the year 2023, based on data provided in a CSV file.
