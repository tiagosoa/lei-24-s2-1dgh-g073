# US015 - Monthly Water Consumption

## INTRODUCTION
The goal of this user story (US15) is to predict the average monthly cost associated with water consumption for a new park with an area of 55 hectares. To achieve this, we will develop a linear regression model that uses park area as the independent variable and the average monthly water consumption cost as the dependent variable. This model will help us understand how water consumption costs scale with park size and provide a cost estimate for the new park.

The cost of water consumption follows specific rules:

1. Up to 1000 cubic meters (m³) per month: 0.7 AC/m³.

2. Above 1000 cubic meters (m³) per month: an additional fee of 15% is added to the base rate of 0.7 AC/m³ for the excess consumption.

To obtain the linear regression model, we will apply linear regression using stats.linregress to model the relationship between park area and average monthly water consumption cost.

We will utilize two primary data sources for this analysis:

1. `Area.csv`: This file contains the area (in hectares) of various parks. The data includes the following columns:
        - Park: Name of the park.
        - Area: Area of the park in hectares.
        
2. `water_consumption_updated.csv`: This file records the daily water consumption (in cubic meters, m³) for each park since the day each park opened. The data includes the following columns:
        - Park: Name of the park.
        - Year: Year of the recorded water consumption.
        - Month: Month of the recorded water consumption.
        - Day: Day of the recorded water consumption.
        - Consumption: Daily water consumption in cubic meters (m³).

```python
import os
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


area_df = pd.read_csv('data/Area.csv', delimiter=';')
water_consumption_df = pd.read_csv('data/water_consumption_updated.csv', delimiter=';')

print("Area Data:")
print(area_df.head())
print("\nWater Consumption Data:")
print(water_consumption_df.head())


area_df['Area'] = area_df['Area'].str.replace(',', '.').astype(float)
water_consumption_df['Consumption'] = water_consumption_df['Consumption'].str.replace(',', '.').astype(float)


water_consumption_df['Date'] = pd.to_datetime(water_consumption_df[['Year', 'Month', 'Day']])


water_consumption_df['Month'] = water_consumption_df['Date'].dt.to_period('M')
monthly_consumption = water_consumption_df.groupby(['Park', 'Month'])['Consumption'].sum().reset_index()
monthly_consumption['Month'] = monthly_consumption['Month'].dt.to_timestamp()


def calculate_monthly_cost(consumption):
    if consumption <= 1000:
        return consumption * 0.7
    else:
        excess = consumption - 1000
        return (1000 * 0.7) + (excess * 0.7 * 1.15)

monthly_consumption['Cost'] = monthly_consumption['Consumption'].apply(calculate_monthly_cost)


average_monthly_cost = monthly_consumption.groupby('Park')['Cost'].mean().reset_index()


merged_df = pd.merge(average_monthly_cost, area_df, on='Park')
merged_df.rename(columns={'Cost': 'AverageMonthlyCost'}, inplace=True)

slope, intercept, r_value, p_value, std_err = stats.linregress(merged_df['Area'], merged_df['AverageMonthlyCost'])


r_squared = r_value ** 2

predicted_cost = intercept + slope * 55
print(f'The average monthly cost is: {predicted_cost:.2f}€')
print(f'R^2 value: {r_squared:.4f}')


plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Area'], merged_df['AverageMonthlyCost'], color='blue', label='Data points')
plt.xlabel('Park Area (hectares)')
plt.ylabel('Average Monthly Cost (AC)')
plt.title('Scatter Plot: Park Area vs. Average Monthly Cost')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Area'], merged_df['AverageMonthlyCost'], color='blue', label='Data points')
plt.plot(merged_df['Area'], intercept + slope * merged_df['Area'], color='red', label='Regression line')
plt.xlabel('Park Area (hectares)')
plt.ylabel('Average Monthly Cost (AC)')
plt.title('Linear Regression: Park Area vs. Average Monthly Cost')
plt.legend()
plt.show()

```

## ANALYSIS

Using the regression model, we predicted the average monthly cost for a 55-hectare park. The predicted cost was approximately 1934.17€.

### Cost Calculation

Considering that the amount paid for water is 0.7€/m³ up to a consumption of 1000 m³/month, with a fee of 15% added for higher consumption levels, we can now calculate the Average Monthly Cost:

### Calculation Steps

Monthly Consumption Calculation:

The monthly consumption for each park is calculated by summing the daily consumption values for each month. This is done as follows:
    
$$
Monthly\;Consumption = \sum_{i=1}^{i = last\;day\;of\;the\;month}\;Daily\;Consumption
$$


Monthly Cost Calculation:

$$
Cost = \left\{\begin{matrix}
Consumption*0,7\;if\;Consumption\;\leq\;1000m^3 \\ 
(1000*0,7)\;+\;(Excess*0,7*1.15)\;if\;Consumption\;>\;1000m^3 
\end{matrix}\right.
$$

where:
$$
Excess=Consumption−1000
$$


Average Monthly Cost:

$$
Average\;Monthly\;Cost =\frac{1}{N}\sum_{i=1}^{N}\; Monthly\;Cost
$$

Linear Regression Model:
$$
Average\;Monthly\;Cost=Intercept+(Slope*Area)
$$

### Linear Regression Analysis

The scatter plot and the regression line show a positive correlation between park area and average monthly cost. The data points generally follow the trend of the regression line.

Slope: Indicates the increase in average monthly cost for each additional hectare of park area.

Intercept: Represents the baseline cost when the park area is zero.

R² Value: The proportion of the variance in the dependent variable that is predictable from the independent variable(s).


For the given data, the regression analysis results in the following values:

Slope: The coefficient representing the rate of change of the average monthly cost with respect to park area.

Intercept: The fixed cost when the park area is zero.

R² value: Measures the goodness of fit of the regression model.

### Benefits of Analyzing the Monthly Water Cost for the Park

This analysis benefits park management by enabling better financial planning and resource allocation. For instance, knowing that the 55-hectare park will have a net monthly spend of approximately 1934.17€ on water allows for more accurate budgeting and potential cost-saving measures.

# US016 - Algorithm Test Size and Execution Time

## INTRODUCTION
The goal of this user story is to analyse the results obtained in US14, and after execute a more advanced analysis through a polynimial regression. In US14 the main objective was to run tests for inputs of variable size, to observe the asymptotic behavior of the execution time of the US13 algorithm. Now in US16 we pretend to dive deeper in the relationship between the  size of the tests and their execution time.

By applying polynomial regression to this data, we can determine the best-fitting curve and gain valuable insights into the algorithm's performance characteristics.

We have one data source for this User Story:

1. `solution_us14.csv`


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


data = pd.read_csv("data/solution_us14.csv", delimiter=";")
data['time'] = data['time'].str.replace(',', '.').astype(float)
X = data['size'].values.reshape(-1, 1)
y = data['time'].values

plt.scatter(X, y, color='blue')
plt.xlabel('Size')
plt.ylabel('Time')
plt.title('Scatter plot of Size vs Time')
plt.show()

def plot_polynomial_regression(X, y, degree):
    polynomial_features = PolynomialFeatures(degree=degree)
    X_poly = polynomial_features.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    y_poly_pred = model.predict(X_poly)
    
    plt.scatter(X, y, color='blue')
    plt.plot(X, y_poly_pred, color='red')
    plt.xlabel('Size')
    plt.ylabel('Time')
    plt.title(f'Polynomial Regression')
    plt.show()
    
    r2 = r2_score(y, y_poly_pred)
    print(f'R^2: {r2}')
    
    return model


degree = 2
model = plot_polynomial_regression(X, y, degree)
```

## ANALYSIS

### Code Analysis:

Data Loading: The code begins by loading the data from the CSV file "solution_us14.csv" using Pandas.

Data Preprocessing: The time column is processed to replace commas with periods and converted to float type. The data is then split into the feature (X) and target (y) variables.

Polynomial Regression Function: The function plot_polynomial_regression() is defined to fit a polynomial regression model of a specified degree to the data. It uses PolynomialFeatures from Scikit-learn to transform the input feature into polynomial features and then fits a linear regression model. It plots the original data points along with the regression line.

The polynomial regression model can be represented as:

$$
y=β_{0}+β_{1}x+β_{2}x^{2}+…+β_{n}x^{n}
$$

Graph Plotting: The code plots the original data points using plt.scatter(), then fits and plots polynomial regression lines for degrees 1 to 4 using the defined function. It also prints the R-squared values for each degree.

Additional Plotting: The code plots a separate graph showing only the data points (dots) without the regression line.

### Results Analysis:

Only the model for Degree 2 is shown, but we will perform an analysis of the models of all four degrees.

Polynomial Regression Analysis:
     Degree 1 (Linear Regression): The linear regression model (degree 1) seems to underfit the data, as it cannot capture the curvature present in the data points.
     Degree 2 (Quadratic Regression): The quadratic regression model (degree 2) provides a better fit to the data compared to the linear model. It captures the curvature in the data more accurately.
     Degrees 3 and 4: These higher-degree polynomial models might suffer from overfitting, especially considering the limited amount of data available. They may capture noise in the data rather than true underlying patterns.

R² Values:
     R² is a measure of how well the regression model fits the observed data. It ranges from 0 to 1, where 1 indicates a perfect fit.
     The R² value is printed for each degree of polynomial regression. It measures the goodness of fit of the model to the data.
     In this case, R² for degree 2 polynomial regression serves as a useful metric to quantify the model's performance.

Data Visualization:
     The scatter plot provides a visual representation of the data distribution.
     The polynomial regression lines overlaid on the scatter plot help visualize how well each degree of polynomial fits the data.

Additional Data Points Plot:
        The additional plot displaying only the dots allows for a clearer view of the distribution of the data without the influence of the regression line.
